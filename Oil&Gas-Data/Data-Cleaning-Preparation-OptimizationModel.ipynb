```python
data_set1 = data_set.iloc[:,-10:]
data_set1['Lateral Length(m)'] = data_set['Lateral Length (m)']
# data_set1 = ll = data_set1.loc[data_set1['Lateral Length (m)'] >= 0]
data_set1
```

```python
data_set1.drop(data_set1.iloc[:, 1:4], inplace = True, axis = 1) 
data_set1 = data_set1.loc[data_set1['Cumulative BOE Prod. (Bbl)'] > 0]
data_set1
```

```python
data_set1.loc[:,'Revenue'] = 20 * data_set1.loc[:,'Cumulative BOE Prod. (Bbl)']
data_set1
```

```python
data_set1.loc[:,'Operational Cost'] = 1460 * data_set1.loc[:,'Lateral Length(m)']
data_set1
```

```python
data_set1.loc[:,'Profit'] = data_set1.loc[:,'Revenue'] - data_set1.loc[:,'Operational Cost']
data_set1
```

```python
print(data_set1['Profit'].max())
print(data_set1['Lateral Length(m)'].max())
print(data_set1['Cumulative BOE Prod. (Bbl)'].max())
```

```python
from ortools.linear_solver import pywraplp


def main():
    # Create the linear solver with the GLOP backend.
    solver = pywraplp.Solver.CreateSolver('GLOP')

    # Create the variables x and y.
    x = solver.NumVar(0, data_set1['Cumulative BOE Prod. (Bbl)'].max(), 'x')
    y = solver.NumVar(0, 4000, 'y')

    print('Number of variables =', solver.NumVariables())

    # Create a linear constraint, 0 <= x + y <= 2.
#     ct = solver.Constraint(62909280, data_set1['Profit'].max(), 'ct')
#     ct.SetCoefficient(x, 1)
#     ct.SetCoefficient(y, 1)
    solver.Add( 20 * x - 1460*y >= 0.0)
    solver.Add( 20*x - 1460*y >= data_set1['Profit'].max())


    print('Number of constraints =', solver.NumConstraints())

    # Create the objective function, 3 * x + y.
#     objective = solver.Objective()
#     objective.SetMaximization()

    solver.Maximize(x + y)
    status = solver.Solve()

#     print('Solution:')
#     print('Objective value =', objective.Value())
    print('x =', x.solution_value())
    print('y =', y.solution_value())
if __name__ == '__main__':
    main()
```
